```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url=BeautifulSoup('https://www.sharesansar.com/live-trading/','html.parser')
soup=requests.get(url)
codes=BeautifulSoup(soup.text,'lxml')
```

```python
data=[]
codes_table=codes.table
tages=codes_table.find_all('tr')
for i in tages:
    data.append(i.text.split('\n'))
    
data=list(filter(None,data))
data
```

```python
for i in range(len(data)):
    data[i]=list(filter(None,data[i]))
    



data[1][0:]
```

```python
finaldata=[]
for i in range(len(data)):
    remove_space = [x.strip(' ') for x in data[i]]
    finaldata.append(remove_space)
finaldata
```

```python

file=open('share.csv','w')
x=csv.writer(file)
x.writerows(finaldata)
file.close()
```

```python
import pandas as pd
df=pd.read_csv('share.csv',encoding='latin1')
df
```

```python
df.dropna()
```

```python
df['LTP']=list(map(lambda x:(x.replace(',','')),df['LTP']))






# df['% Change']=list(map(lambda x:(x.replace(',','')),df['% Change']))

             
df['Open']=list(map(lambda x:(x.replace(',','')),df['Open']))

             
             
df['High']=list(map(lambda x:(x.replace(',','')),df['High']))

             
df['Low']=list(map(lambda x:(x.replace(',','')),df['Low']))


             
df['Prev. Close']=list(map(lambda x:(x.replace(',','')),df['Prev. Close']))
             


```

```python
df['LTP']=list(map(lambda x:float(x),df['LTP']))
df['% Change']=list(map(lambda x:float(x),df['% Change']))
df['Open']=list(map(lambda x:float(x),df['Open']))
df['High']=list(map(lambda x:float(x),df['High']))
df['Low']=list(map(lambda x:float(x),df['Low']))
df['Prev. Close']=list(map(lambda x:float(x),df['Prev. Close']))
df['Point Change']=list(map(lambda x:float(x),df['Point Change']))


```
